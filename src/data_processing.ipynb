```python
import pandas as pd
import numpy as np
```

```python
ext_df=pd.read_csv('../data/extract_dengue_data.csv')
com_df=pd.read_csv('../data/combined_dengue_data.csv')
```

ext_dfの前処理

```python
ext_df.head()
```

```python
ext_df.isnull().sum()
```

```python
ext_df=ext_df.drop('ID',axis=1)
ext_df=ext_df.drop('Unnamed: 0',axis=1)
ext_df['dengue']=1
```

```python
ext_df.head()
```

com_dfの前処理

```python
com_df.head()
```

```python
com_df.isnull().sum()
```

```python
com_df=com_df.dropna(how='all')
com_df=com_df.drop('ID',axis=1)
```

```python
com_df.isnull().sum()
```

```python
dengue_positive_df=com_df[com_df['dengue']==0.0]
dengue_negative_df=com_df[com_df['dengue']==1.0]
```

```python
dengue_positive_df.isnull().sum()
```

```python
dengue_negative_df.isnull().sum()
```

```python
com_df['eye pain'] = com_df['eye pain'].apply(lambda x: np.random.choice([0, 1]) if pd.isna(x) else x)
```

```python
com_df.isnull().sum()
```

```python
com_df=com_df.astype(int)
```
